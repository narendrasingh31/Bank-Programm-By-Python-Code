```python
cust_info = {}
def header():
    h = '-'*80
    bank_name = 'Ethans Bank'
    print(h)
    print(bank_name.center(80))
    print(h)
    
def check_account_exists(account_no):
    if account_no in cust_info:
        return True
    else:
        return False
    
def create_account(account_no,name,password,city):
    global cust_info
    cust_info[account_no] ={'name': name,
                           'password': password,
                           'city': city,
                           'balance': 0
                           }
    if check_account_exists(account_no):
        print(f'Mr. {name} your account has been created successfully ')
        return True
    return False
    
   # print(cust_info)

def login_authentication(account_no, password):
    if check_account_exists(account_no):
        if cust_info[account_no]['password'] == password:
            print('login successfull ')
            account_info(account_no)
        else:
            print('incorrect password  ')
    else:
          print('Account does not exist ')
            
def account_info(account_no):
    while True:
        header()
        print('1.Get Balance \n2.Deposite Money \n3.Withdraw Money \n4.logout')
        option = int(input('Enter your choice : '))
        if option == 1:
            get_balance(account_no)
        if option == 2:
            deposit(account_no)
        if option == 3:
            withdraw(account_no)
        if option ==4:
            break
    
def get_balance(account_no):
    bal = cust_info[account_no]['balance']
    name = cust_info[account_no]['name']
    print(f'Mr. {name} your updated balance is {bal}')
        
def deposit(account_no): 
    amount = int(input('Enter an Amount : '))
    bal = cust_info[account_no]['balance'] + amount
    cust_info[account_no]['balance'] = bal 
    name = cust_info[account_no]['name']
    print(f'Mr/Ms. {name} your updated balance is {bal}')
    
def withdraw(account_no):
    amount = int(input('Enter an Amount : '))
    password = input('Enter your Password : ')
    if cust_info[account_no]['password'] == password:
        if cust_info[account_no]['balance'] >= amount:
            bal = cust_info[account_no]['balance'] - amount
            name = cust_info[account_no]['name']
            cust_info[account_no]['balance'] = bal 
            print(f'Mr.Ms/ {name} your updated balance is {bal}')
        else:
            print('insufficent balance')
    else:
        print('Incorrect password')
        
    
# def verify_account(account_no,password):
#     if check_account_exists(account_no):
#         if cust_info[account_no]['password'] == password:
#             print('login successfull ')
#         else:
#             print('incorrect password  ')
#     else:
#           print('Account doesenot exist ')

while True:
    header()
    print('To login to your account enter 1\nTo create an account enter 2')
    option = int(input('Enter your choice : ')) 
    if option == 2:
        while True:  
            # Account number only have 8 digits and as well as only digits
            account_no = input('Enter your Account number : ')
            if account_no.isdigit() and len(account_no) == 8: 
                break
            else:
                print('ERROR !! Account number can only be digits and only have 8 digits')
        while True:
            name = input('Enter your name : ')
            if not name.isalpha():
                print('ERROR !! Account holder name can only be Alphabet')
            else:
                break
        city = input('Enter your city : ')
        while True:
            # password should have lenght of 4 character
            password = input('Enter your password : ')         
            if not len(password) >= 4:
                print('Invalid Password !! Password should have minimum 4 character long')
            else:
                break
        create_account(account_no, name, password, city)
    
    if option == 1:
        account_no = input('Enter your Account number : ')
        password = input('Enter your password : ')
        login_authentication(account_no, password)
    
```
